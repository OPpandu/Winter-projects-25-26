```python
import torch
import numpy as np
# ===========================================================
# PART 1: Data Generation (The Exclusion Zone)
# ===========================================================

# ⚠️ INSTRUCTION: Replace 1 with your actual Group Number
GROUP_NUMBER = 4

# Set seed for reproducibility
torch.manual_seed(GROUP_NUMBER)
np.random.seed(GROUP_NUMBER)

def generate_data(n_samples=3000):
    """
    Generates 3000 sensor readings for the Kepler-186f Exclusion Zone.
    Shape: Concentric circles (Donut).
    Logic:
      - Dead Zone (1): 2km < distance < 4km
      - Safe Zone (0): distance < 2km OR distance > 4km
    """
    # Generate random coordinates between -5 and 5 km
    X = (torch.rand(n_samples, 2) * 10) - 5

    # Calculate distance from center (radius)
    radius = torch.sqrt(X[:, 0]**2 + X[:, 1]**2)

    # Assign Labels: 1 if inside the Dead Zone, 0 otherwise
    y = ((radius > 2) & (radius < 4)).float().view(-1, 1)

    # Add 5% Noise (Malfunctioning Drones)
    n_noise = int(0.05 * n_samples)
    noise_indices = torch.randperm(n_samples)[:n_noise]
    y[noise_indices] = 1 - y[noise_indices] # Flip labels

    return X, y

# Generate the dataset
X_full, y_full = generate_data(3000)

print(f"Data Generated: {X_full.shape} samples.")
print(f"Target Generated: {y_full.shape} labels.")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(
    X_full, y_full,
    test_size=0.3,
    random_state=GROUP_NUMBER
)
```

```python
def init_params(in_dim, out_dim):
    W = torch.randn(in_dim, out_dim) * torch.sqrt(torch.tensor(2.0 / in_dim))
    b = torch.zeros(out_dim)
    W.requires_grad_()
    b.requires_grad_()
    return W, b
W1, b1 = init_params(2, 32)
W2, b2 = init_params(32, 32)
W3, b3 = init_params(32, 32)
W4, b4 = init_params(32, 1)
```

```python
def tanh(x):
    return torch.tanh(x)
def sigmoid(x):
    return 1 / (1 + torch.exp(-x))
def forward(X):
    a1 = tanh(X @ W1 + b1)
    a2 = tanh(a1 @ W2 + b2)
    a3 = tanh(a2 @ W3 + b3)
    y_hat = sigmoid(a3 @ W4 + b4)
    return y_hat
```

```python
def bce_loss(y_hat, y):
    eps = 1e-7
    y_hat = torch.clamp(y_hat, eps, 1 - eps)
    return -(y * torch.log(y_hat) + (1 - y) * torch.log(1 - y_hat)).mean()
```

```python
def accuracy(y_hat, y):
    return ((y_hat >= 0.5) == y).float().mean().item()
```

```python
epochs = 1000
lr = 0.003
train_losses, val_losses = [], []
train_accs, val_accs = [], []
params = [W1, b1, W2, b2, W3, b3, W4, b4]
for epoch in range(epochs):
    # Forward
    y_hat = forward(X_train)
    loss = bce_loss(y_hat, y_train)
    # Backward
    loss.backward()
    # Manual SGD
    with torch.no_grad():
        for p in params:
            p -= lr * p.grad
            p.grad.zero_()
    # Metrics
    train_losses.append(loss.item())
    train_accs.append(accuracy(y_hat, y_train))
    with torch.no_grad():
        y_val_hat = forward(X_val)
        val_losses.append(bce_loss(y_val_hat, y_val).item())
        val_accs.append(accuracy(y_val_hat, y_val))
    if epoch % 200 == 0:
        print(f"Epoch {epoch} | Loss: {loss.item():.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
# Loss
axs[0].plot(train_losses, label="Train Loss")
axs[0].plot(val_losses, label="Val Loss")
axs[0].set_title("Loss vs Epochs")
axs[0].legend()
# Accuracy
axs[1].plot(train_accs, label="Train Acc")
axs[1].plot(val_accs, label="Val Acc")
axs[1].set_title("Accuracy vs Epochs")
axs[1].legend()
# Decision Boundary
xx, yy = np.meshgrid(
    np.linspace(-5, 5, 300),
    np.linspace(-5, 5, 300)
)
grid = torch.tensor(
    np.c_[xx.ravel(), yy.ravel()],
    dtype=torch.float32
)
with torch.no_grad():
    zz = forward(grid).reshape(xx.shape)
axs[2].contourf(xx, yy, zz, levels=30, cmap="coolwarm", alpha=0.7)
axs[2].scatter(
    X_val[:, 0], X_val[:, 1],
    c=y_val[:, 0], s=10
)
axs[2].set_title("Decision Boundary (Donut)")
plt.show()
```
