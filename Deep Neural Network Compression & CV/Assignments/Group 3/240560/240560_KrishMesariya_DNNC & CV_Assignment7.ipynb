```python
import torch
import numpy as np
# ===========================================================
# PART 1: Data Generation (The Exclusion Zone)
# ===========================================================

# ⚠️ INSTRUCTION: Replace 1 with your actual Group Number
GROUP_NUMBER = 3

# Set seed for reproducibility
torch.manual_seed(GROUP_NUMBER)
np.random.seed(GROUP_NUMBER)

def generate_data(n_samples=3000):
    """
    Generates 3000 sensor readings for the Kepler-186f Exclusion Zone.
    Shape: Concentric circles (Donut).
    Logic:
      - Dead Zone (1): 2km < distance < 4km
      - Safe Zone (0): distance < 2km OR distance > 4km
    """
    # Generate random coordinates between -5 and 5 km
    X = (torch.rand(n_samples, 2) * 10) - 5

    # Calculate distance from center (radius)
    radius = torch.sqrt(X[:, 0]**2 + X[:, 1]**2)

    # Assign Labels: 1 if inside the Dead Zone, 0 otherwise
    y = ((radius > 2) & (radius < 4)).float().view(-1, 1)

    # Add 5% Noise (Malfunctioning Drones)
    n_noise = int(0.05 * n_samples)
    noise_indices = torch.randperm(n_samples)[:n_noise]
    y[noise_indices] = 1 - y[noise_indices] # Flip labels

    return X, y

# Generate the dataset
X_full, y_full = generate_data(3000)

print(f"Data Generated: {X_full.shape} samples.")
print(f"Target Generated: {y_full.shape} labels.")
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

Train / Validation Split

```python
X_train, X_val, y_train, y_val = train_test_split(
    X_full, y_full, test_size=0.3, random_state=GROUP_NUMBER
)
```

MLP Parameter Initialization

```python
def init_params(in_dim, out_dim):
    W = torch.randn(in_dim, out_dim) * 0.1
    b = torch.zeros(1, out_dim)
    W.requires_grad = True
    b.requires_grad = True
    return W, b

W1, b1 = init_params(2, 16)
W2, b2 = init_params(16, 16)
W3, b3 = init_params(16, 16)
W4, b4 = init_params(16, 1)

params = [W1, b1, W2, b2, W3, b3, W4, b4]
```

Activation Functions

```python
def relu(x):
    return torch.clamp(x, min=0)

def sigmoid(x):
    return 1 / (1 + torch.exp(-x))
```

Forward Pass

```python
def forward(X):
    z1 = torch.matmul(X, W1) + b1
    a1 = relu(z1)

    z2 = torch.matmul(a1, W2) + b2
    a2 = relu(z2)

    z3 = torch.matmul(a2, W3) + b3
    a3 = relu(z3)

    z4 = torch.matmul(a3, W4) + b4
    y_hat = sigmoid(z4)

    return y_hat
```

Binary Cross Entropy Loss (Manual)

```python
def binary_cross_entropy(y_pred, y_true):
    eps = 1e-7
    y_pred = torch.clamp(y_pred, eps, 1 - eps)
    loss = - (y_true * torch.log(y_pred) +
              (1 - y_true) * torch.log(1 - y_pred))
    return loss.mean()
```

Accuracy Function

```python
def accuracy(y_pred, y_true):
    preds = (y_pred >= 0.5).float()
    return (preds == y_true).float().mean().item()
```

Training Loop (Manual SGD)

```python
epochs = 300
learning_rate = 0.01

train_losses, val_losses = [], []
train_accs, val_accs = [], []

for epoch in range(epochs):

    # ---- Training ----
    y_pred_train = forward(X_train)
    loss_train = binary_cross_entropy(y_pred_train, y_train)
    acc_train = accuracy(y_pred_train, y_train)

    loss_train.backward()

    with torch.no_grad():
        for p in params:
            p -= learning_rate * p.grad
            p.grad.zero_()

    # ---- Validation ----
    y_pred_val = forward(X_val)
    loss_val = binary_cross_entropy(y_pred_val, y_val)
    acc_val = accuracy(y_pred_val, y_val)

    train_losses.append(loss_train.item())
    val_losses.append(loss_val.item())
    train_accs.append(acc_train)
    val_accs.append(acc_val)

    if epoch % 50 == 0:
        print(f"Epoch {epoch}: "
              f"Train Loss={loss_train.item():.4f}, "
              f"Val Loss={loss_val.item():.4f}")
```

GDC Dashboard (Loss & Accuracy)

```python
plt.figure(figsize=(14,5))

# Loss Plot
plt.subplot(1,2,1)
plt.plot(train_losses, label="Training Loss")
plt.plot(val_losses, label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("System Error Trajectory")
plt.legend()

# Accuracy Plot
plt.subplot(1,2,2)
plt.plot(train_accs, label="Training Accuracy")
plt.plot(val_accs, label="Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Integrity Check")
plt.legend()

plt.show()
```

Decision Boundary Visualization (Final Proof)

```python
# Create mesh grid
x_min, x_max = -5.5, 5.5
y_min, y_max = -5.5, 5.5

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 300),
    np.linspace(y_min, y_max, 300)
)

grid = torch.tensor(np.c_[xx.ravel(), yy.ravel()], dtype=torch.float32)
with torch.no_grad():
    probs = forward(grid).reshape(xx.shape)

plt.figure(figsize=(7,7))

# Decision boundary
plt.contourf(xx, yy, probs, levels=[0,0.5,1], alpha=0.3)

# Validation points
plt.scatter(X_val[:,0], X_val[:,1],
            c=y_val.squeeze(),
            cmap="coolwarm",
            edgecolors="k",
            s=15)

plt.title("Geospatial Threat Map (Decision Boundary)")
plt.xlabel("X Coordinate (km)")
plt.ylabel("Y Coordinate (km)")
plt.show()
```
