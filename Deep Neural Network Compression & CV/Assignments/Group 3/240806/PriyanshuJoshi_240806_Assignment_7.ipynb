$$Loss = -\frac{1}{N} \sum_{i=1}^{N} [y_i \cdot \log(\hat{y}_i) + (1-y_i) \cdot \log(1-\hat{y}_i)]$$

```python
import torch
import numpy as np
# ===========================================================
# PART 1: Data Generation (The Exclusion Zone)
# ===========================================================

# ⚠️ INSTRUCTION: Replace 1 with your actual Group Number
GROUP_NUMBER = 3

# Set seed for reproducibility
torch.manual_seed(GROUP_NUMBER)
np.random.seed(GROUP_NUMBER)

def generate_data(n_samples=3000):
    """
    Generates 3000 sensor readings for the Kepler-186f Exclusion Zone.
    Shape: Concentric circles (Donut).
    Logic:
      - Dead Zone (1): 2km < distance < 4km
      - Safe Zone (0): distance < 2km OR distance > 4km
    """
    # Generate random coordinates between -5 and 5 km
    X = (torch.rand(n_samples, 2) * 10) - 5

    # Calculate distance from center (radius)
    radius = torch.sqrt(X[:, 0]**2 + X[:, 1]**2)

    # Assign Labels: 1 if inside the Dead Zone, 0 otherwise
    y = ((radius > 2) & (radius < 4)).float().view(-1, 1)

    # Add 5% Noise (Malfunctioning Drones)
    n_noise = int(0.05 * n_samples)
    noise_indices = torch.randperm(n_samples)[:n_noise]
    y[noise_indices] = 1 - y[noise_indices] # Flip labels

    return X, y

# Generate the dataset
X_full, y_full = generate_data(3000)

print(f"Data Generated: {X_full.shape} samples.")
print(f"Target Generated: {y_full.shape} labels.")
```

```python
# ==============================================================================
# Code from here. Best of luckk :)
# ==============================================================================
# ===========================================================
# PART 2: Train–Validation Split (70–30)
# ===========================================================

perm = torch.randperm(len(X_full))
split = int(0.7 * len(X_full))

train_idx = perm[:split]
val_idx = perm[split:]

X_train, y_train = X_full[train_idx], y_full[train_idx]
X_val, y_val = X_full[val_idx], y_full[val_idx]


# ===========================================================
# PART 3: Manual MLP Initialization
# ===========================================================

def init_layer(in_dim, out_dim):
    W = torch.randn(in_dim, out_dim) * 0.1
    b = torch.zeros(out_dim)
    W.requires_grad_()
    b.requires_grad_()
    return W, b

W1, b1 = init_layer(2, 16)
W2, b2 = init_layer(16, 16)
W3, b3 = init_layer(16, 16)
W4, b4 = init_layer(16, 1)

params = [W1, b1, W2, b2, W3, b3, W4, b4]


# ===========================================================
# PART 4: Activation Functions
# ===========================================================

def relu(x):
    return torch.clamp(x, min=0)

def sigmoid(x):
    return 1 / (1 + torch.exp(-x))


# ===========================================================
# PART 5: Forward Pass
# ===========================================================

def forward(X):
    z1 = X @ W1 + b1
    a1 = relu(z1)

    z2 = a1 @ W2 + b2
    a2 = relu(z2)

    z3 = a2 @ W3 + b3
    a3 = relu(z3)

    z4 = a3 @ W4 + b4
    return sigmoid(z4)


# ===========================================================
# PART 6: Manual Binary Cross Entropy Loss
# ===========================================================

def binary_cross_entropy(y_hat, y):
    eps = 1e-7
    y_hat = torch.clamp(y_hat, eps, 1 - eps)
    return -torch.mean(
        y * torch.log(y_hat) + (1 - y) * torch.log(1 - y_hat)
    )


# ===========================================================
# PART 7: Training Loop (Manual SGD)
# ===========================================================

lr = 0.05
epochs = 500

train_losses, val_losses = [], []
train_accs, val_accs = [], []

for epoch in range(epochs):

    # ---- Forward
    y_pred = forward(X_train)
    loss = binary_cross_entropy(y_pred, y_train)

    # ---- Backward
    loss.backward()

    # ---- Manual SGD update
    with torch.no_grad():
        for p in params:
            p -= lr * p.grad
            p.grad.zero_()

    # ---- Metrics
    with torch.no_grad():
        val_pred = forward(X_val)
        val_loss = binary_cross_entropy(val_pred, y_val)

        train_acc = ((y_pred > 0.5) == y_train).float().mean()
        val_acc = ((val_pred > 0.5) == y_val).float().mean()

    train_losses.append(loss.item())
    val_losses.append(val_loss.item())
    train_accs.append(train_acc.item())
    val_accs.append(val_acc.item())

    if epoch % 50 == 0:
        print(
            f"Epoch {epoch:03d} | "
            f"Train Loss: {loss:.4f} | Val Loss: {val_loss:.4f} | "
            f"Train Acc: {train_acc:.3f} | Val Acc: {val_acc:.3f}"
        )


# ===========================================================
# PART 8: GDC HUD — Loss, Accuracy, Decision Boundary
# ===========================================================

import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# ---- Loss Plot
axs[0].plot(train_losses, label="Train Loss")
axs[0].plot(val_losses, label="Val Loss")
axs[0].set_title("System Error Trajectory")
axs[0].legend()

# ---- Accuracy Plot
axs[1].plot(train_accs, label="Train Accuracy")
axs[1].plot(val_accs, label="Val Accuracy")
axs[1].set_title("Integrity Check")
axs[1].legend()

# ---- Decision Boundary
xx, yy = np.meshgrid(
    np.linspace(-5, 5, 300),
    np.linspace(-5, 5, 300)
)
grid = torch.tensor(
    np.c_[xx.ravel(), yy.ravel()],
    dtype=torch.float32
)

with torch.no_grad():
    zz = forward(grid).reshape(xx.shape)

axs[2].contourf(xx, yy, zz, levels=[0, 0.5, 1], alpha=0.3)
axs[2].scatter(
    X_val[:, 0], X_val[:, 1],
    c=y_val.squeeze(),
    cmap="coolwarm",
    s=10
)
axs[2].set_title("Geospatial Threat Map")

plt.tight_layout()
plt.show()
```
