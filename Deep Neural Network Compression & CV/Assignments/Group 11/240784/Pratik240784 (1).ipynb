```python
#MLP from scratch

import torch
import numpy as np
import matplotlib.pyplot as plt


num = 11
torch.manual_seed(num)
np.random.seed(num)


# Data Generation

def generate_data(n_samples=3000):
    X = (torch.rand(n_samples, 2) * 10) - 5
    r = torch.sqrt(X[:, 0]**2 + X[:, 1]**2)
    y = ((r > 2) & (r < 4)).float().view(-1, 1)

    # 5% noise
    n_noise = int(0.05 * n_samples)
    idx = torch.randperm(n_samples)[:n_noise]
    y[idx] = 1 - y[idx]

    return X, y

X_raw, y_full = generate_data(3000)


# Feature Engineering

x = X_raw[:, 0:1]
y = X_raw[:, 1:2]
r = torch.sqrt(x**2 + y**2)

X_full = torch.cat([
    x, y,
    r,
    r**2,
    r**3,
    x**2,
    y**2,
    x**3,
    y**3
], dim=1)


# Train & Validation Split

indices = torch.randperm(X_full.shape[0])
train_size = int(0.7 * X_full.shape[0])

X_train = X_full[indices[:train_size]]
y_train = y_full[indices[:train_size]]

X_val = X_full[indices[train_size:]]
y_val = y_full[indices[train_size:]]


# Normalize Features

mean = X_train.mean(dim=0, keepdim=True)
std = X_train.std(dim=0, keepdim=True) + 1e-8

X_train = (X_train - mean) / std
X_val   = (X_val - mean) / std


# MLP Architecture

input_dim = X_train.shape[1]
hidden_dim = 16
output_dim = 1

W1 = torch.randn(input_dim, hidden_dim) * np.sqrt(2 / input_dim)
W1.requires_grad_()

W2 = torch.randn(hidden_dim, hidden_dim) * np.sqrt(2 / hidden_dim)
W2.requires_grad_()

W3 = torch.randn(hidden_dim, hidden_dim) * np.sqrt(2 / hidden_dim)
W3.requires_grad_()

W4 = torch.randn(hidden_dim, output_dim) * np.sqrt(2 / hidden_dim)
W4.requires_grad_()

b1 = torch.ones((1, hidden_dim)) * 0.01
b1.requires_grad_()

b2 = torch.ones((1, hidden_dim)) * 0.01
b2.requires_grad_()

b3 = torch.ones((1, hidden_dim)) * 0.01
b3.requires_grad_()

b4 = torch.zeros((1, output_dim), requires_grad=True)

def relu(z):
    return torch.clamp(z, min=0)

def sigmoid(z):
    return 1 / (1 + torch.exp(-z))

def forward(X):
    a1 = relu(torch.matmul(X, W1) + b1)
    a2 = relu(torch.matmul(a1, W2) + b2)
    a3 = relu(torch.matmul(a2, W3) + b3)
    return sigmoid(torch.matmul(a3, W4) + b4)


# Loss & Accuracy

def binary_cross_entropy(y_hat, y):
    eps = 1e-7
    y_hat = torch.clamp(y_hat, eps, 1 - eps)
    return -torch.mean(y * torch.log(y_hat) + (1 - y) * torch.log(1 - y_hat))

def accuracy(y_hat, y):
    return ((y_hat >= 0.5).float() == y).float().mean()

# Training Loop

learning_rate = 0.05
epochs = 400

params = [W1, b1, W2, b2, W3, b3, W4, b4]

train_losses, val_losses = [], []
train_accs, val_accs = [], []

for epoch in range(epochs):

    y_pred = forward(X_train)
    loss = binary_cross_entropy(y_pred, y_train)
    acc = accuracy(y_pred, y_train)

    loss.backward()

    with torch.no_grad():
        for p in params:
            p -= learning_rate * p.grad
            p.grad.zero_()

    with torch.no_grad():
        val_pred = forward(X_val)
        val_loss = binary_cross_entropy(val_pred, y_val)
        val_acc = accuracy(val_pred, y_val)

    train_losses.append(loss.item())
    val_losses.append(val_loss.item())
    train_accs.append(acc.item())
    val_accs.append(val_acc.item())



# Loss, Accuracy, Donut Plots

fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Loss plot
axs[0].plot(train_losses, label="Train Loss")
axs[0].plot(val_losses, label="Val Loss")
axs[0].set_title("Loss vs Epochs")
axs[0].legend()

# Accuracy plot

axs[1].plot(train_accs, label="Train Acc")
axs[1].plot(val_accs, label="Val Acc")
axs[1].set_title("Accuracy vs Epochs")
axs[1].legend()

# Decision Boundary
xx, yy = np.meshgrid(np.linspace(-5, 5, 300),
                     np.linspace(-5, 5, 300))

xg = torch.tensor(xx.reshape(-1,1), dtype=torch.float32)
yg = torch.tensor(yy.reshape(-1,1), dtype=torch.float32)
rg = torch.sqrt(xg**2 + yg**2)

Xg = torch.cat([
    xg, yg,
    rg,
    rg**2,
    rg**3,
    xg**2,
    yg**2,
    xg**3,
    yg**3
], dim=1)

Xg = (Xg - mean) / std

with torch.no_grad():
    Z = forward(Xg).reshape(xx.shape)

axs[2].contourf(xx, yy, Z, levels=[0, 0.5, 1], alpha=0.6)
axs[2].scatter(X_raw[indices[train_size:], 0],
               X_raw[indices[train_size:], 1],
               c=y_val.squeeze(), s=10, edgecolors='k')
axs[2].set_title("Decision Boundary")

plt.show()
```
